```python
import pandas as pd
import dash
import plotly.express as px
import plotly.graph_objects as go
from dash import dcc, html
from dash.dependencies import Input, Output
import dash_bootstrap_components as dbc
```

```python
fidf = pd.read_excel("finance_data.xlsx") # Financial Team DataFrame
badf = pd.read_excel("ba_data.xlsx") # BA Team DataFrame
```

```python
app = dash.Dash(__name__,external_stylesheets=[dbc.themes.BOOTSTRAP])
```

```python
app.layout = html.Div([
     html.Div([
         html.Div([
        dcc.Graph(id='the_graph',figure = {})
        ]),
     html.Label(['Choose the no of Managers:'],style={'font-weight': 'bold', "text-align": "left"}),
     dcc.Dropdown(id='manager_ba',
            options=[{'label':x, 'value':x} for x in range(100)],
            value=1,
            multi=False,
            disabled=False,
            clearable=False,
            searchable=True,
            placeholder='Choose the no of Managers',
            className='form-dropdown',
            style={'width':"10%"},
            persistence='number',
            persistence_type='memory'),
    html.Label(['Choose the no of Fresher - Team leader:'],style={'font-weight': 'bold', "text-align": "left"}),
    dcc.Dropdown(id='tl',
            options=[{'label':x, 'value':x} for x in range(100)],
            value=2,
            multi=False,
            disabled=False,
            clearable=False,
            searchable=True,
            placeholder='Choose the no of Fresher - Team leader',
            className='form-dropdown',
            style={'width':"10%"},
            persistence='number',
            persistence_type='memory'),
    html.Label(['Choose the no of Fresher - Team leader (RA):'],style={'font-weight': 'bold', "text-align": "left"}),
    dcc.Dropdown(id='ra_tl',
            options=[{'label':x, 'value':x} for x in range(10)],
            value=2,
            multi=False,
            disabled=False,
            clearable=False,
            searchable=True,
            placeholder='Choose the no of Fresher - Team leader (A)',
            className='form-dropdown',
            style={'width':"10%"},
            persistence='number',
            persistence_type='memory'),
    html.Label(['Choose the no of Fresher - Team leader (DA):'],style={'font-weight': 'bold', "text-align": "left"}),
    dcc.Dropdown(id='da_tl',
            options=[{'label':x, 'value':x} for x in range(100)],
            value=2,
            multi=False,
            disabled=False,
            clearable=False,
            searchable=True,
            placeholder='Choose the no of Fresher - Team Leader (DA)',
            className='form-dropdown',
            style={'width':"10%"},
            persistence='number',
            persistence_type='memory'),
    html.Label(['Choose the no of Contingency team memebers:'],style={'font-weight': 'bold', "text-align": "left"}),
    dcc.Dropdown(id='ct',
            options=[{'label':x, 'value':x} for x in range(100)],
            value=5,
            multi=False,
            disabled=False,
            clearable=False,
            searchable=True,
            placeholder='Choose the no of Contingency team memebers',
            className='form-dropdown',
            style={'width':"10%"},
            persistence='number',
            persistence_type='memory'),
    html.Label(['Choose the no of Interns:'],style={'font-weight': 'bold', "text-align": "left"}),
    dcc.Dropdown(id='interns',
            options=[{'label':x, 'value':x} for x in range(100)],
            value=24,
            multi=False,
            disabled=False,
            clearable=False,
            searchable=True,
            placeholder='Choose the no of Interns',
            className='form-dropdown',
            style={'width':"10%"},
            persistence='number',
            persistence_type='memory')
])
])
```

```python
@app.callback(
    Output(component_id='the_graph', component_property='figure'),
    [Input(component_id='manager_ba', component_property='value'),
     Input(component_id='tl', component_property='value'),
     Input(component_id='ra_tl', component_property='value'),
     Input(component_id='da_tl', component_property='value'),
     Input(component_id='ct', component_property='value'),
     Input(component_id='interns', component_property='value')]
)
 
def final(ma,tl,ra,da,ct,inte):
    fdf = pd.DataFrame()
    lst = [ma,tl,inte,ra,inte,da,inte,ct]
    fdf ["Position"] = badf['Position']
    fdf ["Salary"] = badf["Salary/month/employee"]
    fdf ["noemp"] = lst
    fdf ["values"] = fdf.Salary * fdf.noemp
    
    fig = px.bar(fdf, x='Position', y="values", color='Position')
    return fig
    

if __name__ == '__main__':
    app.run_server(port = 1001)
```
